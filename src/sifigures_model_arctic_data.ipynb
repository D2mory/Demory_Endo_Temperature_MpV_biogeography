```r
## Temperature vector
T = 0:0.1:25
Texp = c(0.5,2.5,3.5,7)
TK = T+273.15
```

```r
##  Model parameters
# Model parameters
para = c(1E11,7260.7,1.998E24,16296,1E10,2.6279e-07,291.83,
         0.1,2.0983E09,5705.3,3.0485E24,16007,99.534,280.5,
         9.2464,5.0254E22,0.49447,0.47675,1.489e-10)
```

```r
## Hyper-parameter functions from Demory Ecology Letters 2021.
Tdrivenf <- function(T,para){
TK = T+273.15
# --- Growth and basal mortality ---
A1 = para[1]
E1 = para[2]
A2 = para[3]
E2 = para[4]
f = A1*exp(-E1/TK)
g = A2*exp(-E2/TK)
mu = f
psi = g
mu_net = f-g
mu_net[mu_net<0]=0
# --- Carrying capacity ---
K = para[5]

# --- Viral adsorption ---
phiK = para[6]
Tphi = para[7]
phir = para[8]
numA = phiK
denA = phiK+exp(-phir*(TK-Tphi))
phi = numA/denA

## --- Lysis rate ---
s1 = para[9]
d1 = para[10]
s2 = para[11]
d2 = para[12]
f = s1*exp(-d1/TK)
g = s2*exp(-d2/TK)
lambda = f - g
lambda[lambda<0] = 0

## --- Burst size ---
b1 = para[13]
beta = b1*(f-g)
beta[beta<0] = 0

## --- production of noninfectious viruses
epsK = 1
Teps = para[14]
epsr = para[15]
epsilon = epsK/(1+exp((-d2/TK)*(1-(Teps/TK)))^epsr)

## --- viral lost of infectivity ---
sig1 = para[16]
sigma = sig1*exp(-d2/TK)

## --- viral degradation ---
#delta parameters
del1 = para[17]
del2 = para[18]
delta = del1*exp(-del2/TK)

# --- viral aggregation ---
omega = para[19]

fT_arc = data.frame("Temperature"=T,"mu"=mu,"psi"=psi,"K"=K,
                    "mu_net"= mu_net,"phi"=phi,
                    "lambda"=lambda,"beta"=beta,"epsilon"=epsilon,"sigma"=sigma,
                    "delta"=delta,"omega"=omega)
return(fT_arc)
}
```

```r
# Arrange the parameter dataframe
fT_arc = Tdrivenf(T,para)
fT_arc$K = c()
fT_arc$omega = c()
fT_arc_plot = melt(fT_arc, id.vars = "Temperature")
```

```r
# ODE model
ODEmodelT = function(t,y,parin){
  ## Definition of the variables
  S = y[1]  # Susceptible Phytoplankton cells 
  I = y[2]  # Infected Phytoplankton cells    
  V = y[3]  # Free virus particles              
  W = y[4]  # Non-infectious virus                          
  
  # Total host population
  N = S+I

  ## Parameters
  theta   = Tdrivenf(parin$T[1],parin$para)
  mu      = theta$mu
  psi     = theta$psi
  K       = theta$K
  phi     = theta$phi
  lambda  = theta$lambda
  epsilon = theta$epsilon
  beta    = theta$beta
  sigma   = theta$sigma
  delta   = theta$delta
  omega   = theta$omega
  
  ## ODE
  dS = mu*S*(1-N/K) - psi*S - phi*V*S
  dI = phi*V*S - psi*I - lambda*I        
  dV = (1-epsilon)*beta*lambda*I - phi*S*V - sigma*V - omega*(V+W)*V         
  dW = epsilon*beta*lambda*I + sigma*V - delta*W - omega*(V+W)*W                 
  
  dy = c(dS,dI,dV,dW)
  return(list(dy))
}
```

```r
## Model integration

# Integration conditions
# Time
t0 = 0
dt = 0.01
tf = 5
t  = seq(t0,tf,dt)

mod.V = c()
mod.M = c()
for (i in 1:4){
 
# Initial variable conditions
S0 = Ministat_mean[i]
I0 = 0
V0 = Vinistat_mean[i]
G0 = 0
y0 = c(S0,I0,V0,G0)

# Parameters
parin = data.frame("para" = para, "T" = Texp[i])
# Integration
ymod=ode(y0,t,ODEmodelT,parin)

Micro = ymod[,2]+ymod[,3]
Vir   = ymod[,4]+ymod[,5]

mod.M = cbind(mod.M,Micro)
mod.V = cbind(mod.V,Vir)
}

# Arrange model outputs
colnames(mod.M) = c("T0.5","T2.5","T3.5","T7")
colnames(mod.V) = c("T0.5","T2.5","T3.5","T7")

df_list_modV = list(data.frame("time"=t,"abundances"=mod.V[,1],
                                "T"=0.5),
                    data.frame("time"=t,"abundances"=mod.V[,2],
                                "T"=2.5),
                    data.frame("time"=t,"abundances"=mod.V[,3],
                                "T"=3.5),
                    data.frame("time"=t,"abundances"=mod.V[,4],
                                "T"=7))
fitV = melt(df_list_modV,id.vars = c("time","T"))

df_list_modM = list(data.frame("time"=t,"abundances"=mod.M[,1],
                               "T"=0.5),
                    data.frame("time"=t,"abundances"=mod.M[,2],
                               "T"=2.5),
                    data.frame("time"=t,"abundances"=mod.M[,3],
                               "T"=3.5),
                    data.frame("time"=t,"abundances"=mod.M[,4],
                               "T"=7))
fitM = melt(df_list_modM,id.vars = c("time","T"))
```
